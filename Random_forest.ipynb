```r
library(ggplot2)

# Load the breast cancer dataset
data <- read.csv('updated_dataset.csv')

# Check the structure of the dataset
str(data)

# Convert the 'diagnosis' variable to a factor with two levels
data$diagnosis <- factor(data$diagnosis, levels = c("M", "B"))

# Create a Linear Regression model
model <- lm(diagnosis ~ ., data = data)

# Make predictions on the test set
y_pred <- predict(model, newdata = data)

# Plot the linear regression line along with data points
ggplot(data, aes(x = texture_mean, y = diagnosis)) +
  geom_point(color = 'blue') +
  geom_smooth(method = 'lm', color = 'red', se = FALSE) +
  labs(x = 'Texture Mean', y = 'Diagnosis') +
  ggtitle('Linear Regression')

# Calculate lift
baseline <- mean(as.numeric(data$diagnosis))  # Convert y to numeric before calculating mean
lift <- abs(y_pred - baseline)

# Plot the lift curve
ggplot(data, aes(x = texture_mean, y = lift)) +
  geom_line(color = 'green') +
  labs(x = 'Texture Mean', y = 'Lift') +
  ggtitle('Lift Curve')
```

```r
library(ggplot2)

# Load the breast cancer dataset
data <- read.csv('datascience.csv')

# Convert the 'diagnosis' variable to a factor with two levels
data$diagnosis <- factor(data$diagnosis, levels = c("M", "B"))

# Create a Linear Regression model
model <- lm(diagnosis ~ radius_worst + texture_worst, data = data)

# Make predictions on the test set
y_pred <- predict(model, newdata = data)

# Plot the linear regression line along with data points
ggplot(data, aes(x = texture_worst, y = radius_worst, color = diagnosis)) +
  geom_point() +
  geom_smooth(method = 'lm', se = FALSE) +
  labs(x = 'Texture Worst', y = 'Radius Worst', color = 'Diagnosis') +
  ggtitle('Linear Regression')

# Calculate lift
baseline <- mean(as.numeric(data$diagnosis))  # Convert y to numeric before calculating mean
lift <- abs(as.numeric(y_pred) - baseline)

# Plot the lift curve
ggplot(data, aes(x = texture_worst, y = lift)) +
  geom_line() +
  labs(x = 'Texture Worst', y = 'Lift') +
  ggtitle('Lift Curve')
```

```r
# Load the necessary library
library(randomForest)

# Load the breast cancer dataset
data <- read.csv('datascience.csv')

# Convert the 'diagnosis' variable to a factor with two levels
data$diagnosis <- factor(data$diagnosis, levels = c("M", "B"))

# Split the data into features and target variable
X <- data[, -c(1, 2)]  # Exclude 'id' and 'diagnosis' columns
y <- data$diagnosis

# Train the Random Forest model
set.seed(42)  # For reproducibility
rf_model <- randomForest(diagnosis ~ ., data = data, ntree = 100)

# Print the model summary
print(rf_model)

# Plot variable importance
varImpPlot(rf_model)

# Make predictions on the training data
y_pred <- predict(rf_model, data)

# Evaluate the model
confusion_matrix <- table(y, y_pred)
print(confusion_matrix)

# Calculate accuracy
accuracy <- sum(diag(confusion_matrix)) / sum(confusion_matrix)
print(paste("Accuracy:", accuracy))
```

```r
# Install required packages
install.packages("caret")
install.packages("randomForest")
install.packages("ggplot2")

# Load required libraries
library(caret)        # For data preprocessing and modeling
library(randomForest) # For building a random forest model
library(ggplot2)

# Read the breast cancer dataset
data <- read.csv("updated_dataset.csv", header = TRUE)

# Replace "NA" values with actual NA values in R
data[data == "NA"] <- NA

# Convert categorical variables to factors
# Modify the following lines according to your dataset's column names
data$Diagnosis <- factor(data$Diagnosis)    # Target variable
data$Feature1 <- factor(data$Feature1)      # Categorical predictor 1
data$Feature2 <- factor(data$Feature2)      # Categorical predictor 2
# Continue with the conversion for other categorical predictors

# Handling missing values. For simplicity, we will remove rows with missing values
data_clean <- na.omit(data)

# Split the data into training and testing sets
set.seed(123) # For reproducibility
train_index <- createDataPartition(data_clean$Diagnosis, p = 0.7, list = FALSE)
train_data <- data_clean[train_index, ]
test_data <- data_clean[-train_index, ]

# Define predictor variables and target variable
# Modify the following line to include your dataset's predictor variables
predictors <- c("Feature1", "Feature2", ..., "LastFeature")
target <- "Diagnosis"

# Build a predictive model (Random Forest)
model <- train(train_data[, predictors], train_data[, target], method = "rf")

# Make predictions on the test set
predictions <- predict(model, newdata = test_data[, predictors])

# Evaluate the model
confusion_matrix <- table(predictions, test_data[, target])
accuracy <- sum(diag(confusion_matrix)) / sum(confusion_matrix)

# Print the confusion matrix and accuracy
print(confusion_matrix)
print(paste("Accuracy:", accuracy))

# Convert confusion matrix to data frame
confusion_df <- as.data.frame.table(confusion_matrix)

# Plot confusion matrix. Create confusion matrix plot
barplot(confusion_matrix,
        main = "Confusion Matrix",
        xlab = "Predicted",
        ylab = "Actual",
        col = c("lightblue", "lightgreen"),
        legend = rownames(confusion_matrix),
        beside = TRUE)
```
